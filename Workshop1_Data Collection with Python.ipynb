# 1. Get from MySQL Database 

```python
! pip install pymysql
! pip install pandas


import pymysql
import pandas as pd

ENDPOINT = '127.0.0.1'   #127.0.0.1
PORT = 3306 #3306
USERNAME = 'root'
DBNAME = 'pymysql_demo'
PASSWORD = 'Password1234' # Change this based on your own password
CURSORCLASS = pymysql.cursors.DictCursor

def initiate_local_connection():
    try:
        connection = pymysql.connect(host=ENDPOINT,
                                     port=PORT,
                                     user=USERNAME,
                                     passwd=PASSWORD,
                                     db=DBNAME,
                                     cursorclass=CURSORCLASS
                                     )
        print('[+] Local Connection Successful')
    except Exception as e:
        print(f'[+] Local Connection Failed: {e}')
        connection = None

    return connection

connection = initiate_local_connection()

```

## List Table

```python
# List all tables to look what we have in the table 
# 'show table' command

cursor = connection.cursor()
cursor.execute("show tables;") #หลังexecute จะเป็น ภาษาSQL
tables = cursor.fetchall() 
cursor.close()

print(tables)
```

## Query table

```python
# Using 'with connection.cursor() as cursor:' to query data

with connection.cursor() as cursor:
  cursor.execute("SELECT * FROM employees;")
  result = cursor.fetchall()

print("Number of rows: ", len(result))
```

```python
result
```

```python
type(result)
```

## Convert to Pandas

```python
import pandas as pd
employees = pd.DataFrame(result)
type(employees)
```

```python
employees
```

```python
employees = employees.set_index("EMPLOYEE_ID") #function set_index
employees
```

## Query with Pandas

```python
# Query by using read_sql() 

sql = "SELECT * FROM jobs"
jobs = pd.read_sql(sql, connection)
jobs
```

## Join table : Employees and jobs
- join with Primary key "JOB_ID"

```python
employees = employees.merge(jobs,
                            how = "left",
                            left_on = "JOB_ID",
                            right_on = "JOB_ID")
employees
```

## convert to Pandas

```python
conversion_rate=pd.DataFrame(result_conversion_rate)
conversion_rate
```

```python
# Convert Index (date) to normal column or Reset Index
conversion_rate = conversion_rate.reset_index().rename(columns={"index":"date"})
conversion_rate[:3] #show head 3
```

## Join data

Join data with
- Employees
- conversion_rate

```python
employees[:3] #head 3
```

```python
#Change 'HIRE_DATE' to "date"
# Copy 'HIRE_DATE' to new column named 'date'
employees['date']=employees['HIRE_DATE']
employees.head()
```

```python
# Change to date type of employees and conversion_rate
employees['date']=pd.to_datetime(employees['date']).dt.date
conversion_rate['date']=pd.to_datetime(conversion_rate['date']).dt.date

conversion_rate
```

```python
# We got the same data type then JOIN data !!!
final_df = employees.merge(conversion_rate,
                           how='left',
                          left_on='date',
                          right_on='date')
final_df.head()
```

```python
#Check null in 'conversion_rate' column
final_df['conversion_rate'].isnull().sum()
```

```python
final_df.dtypes
```

```python
#Convert column 'SALARY' to float type 
final_df['SALARY']=final_df['SALARY'].astype(float)
final_df.head(3)
```

```python
#and Check column 'SALARY' again
final_df['SALARY'].dtypes
```

```python
#Create new column 'THB_SALARY'
final_df['THB_SALARY'] = final_df['SALARY']*final_df['conversion_rate']
final_df.head(3)
```

```python
# drop irrelevant column 
final_df=final_df.drop('date',axis=1)
final_df.tail()
```

## Save to CSV file
